## Modelos de ML
#### Objetivo: predecir temperatura

Usando un modelo supervisado, se usaran las caracteristicas latitud, longitud, año y mes, para predecir la temperatura en cada pais, en cada ciudad, en cada año y en cada mes. Antes de comenzar, se deben arreglar aquellas columnas que contienen caracteres no numericos, como latitud y longitud. La primera termina en 'N' (norte) o 'S' (sur) y la segunda termina en 'E' (este) y 'W' (oeste). Utilizando la funcion coord_change, se transforman las letras 'N' y 'E' en +1 y las letras 'S' y 'W' en -1. Este numero multiplica la parte numerica del valor. Finalmente, se obtiene un float a partir de un string.

```python
# string: texto
# float: numero decimal

def coord_change(value):
    """
    input: valor de latitude o longitude (string)
    transforma el caracter final del string (N,S,E o W) a +1 o -1 y lo multiplica por la parte numerica del string
    output: -valor o +valor (float)
    """
    # si value es float no hace cambios
    if isinstance(value, float):
        return value

    # si value es string cambia su valor
    num = float(value[:-1]) # value sin el ultimo caracter
    coord = value[-1] # ultimo caracter de value

    if coord in ['S','W']:
        num = num*(-1)
    else:
        num = num

    return num
    
```

```python
# transformacion de latitude y longitude
MajorCityDF['Latitude']= MajorCityDF['Latitude'].apply(coord_change)
MajorCityDF['Longitude']= MajorCityDF['Longitude'].apply(coord_change)

# Calcula la temperatura promedio por cada mes, año, ciudad y pais
# El promedio de latitud y longitud solo se usa para añadir esos datos a cada grupo
groups = MajorCityDF.groupby(["Country", "City", "year", "month"])[["AverageTemperature","Latitude", "Longitude"]].mean().reset_index()
features = groups[["Latitude", "Longitude", "year", "month"]]

# Se usan la temperatura promedio por mes como etiqueta
labels = groups['AverageTemperature']
```

### 1. Decission tree (DT)

```python
# Dividir en set de entrenamiento y prueba
# 20% del dataset es set de entrenamiento
Xtrain, Xtest, ytrain, ytest = train_test_split(features,
                                                labels,
                                                test_size=0.2,
                                                random_state=100)

# Crear modelo de decission tree
model = DecisionTreeRegressor(random_state=100)

# Evaluar el modelo con todo el set de entrenamiento
model.fit(Xtrain, ytrain)

# Obtener la prediccon del modelo
y_pred = model.predict(Xtest)
```

```python
# Evaluar el modelo con metricas de regresion
print("MAE:", metrics.mean_absolute_error(ytest, y_pred))
print("MSE:", metrics.mean_squared_error(ytest, y_pred))
print("RMSE:", np.sqrt(metrics.mean_squared_error(ytest, y_pred)))
print("R²:", metrics.r2_score(ytest, y_pred))
```

### Curva de aprendizaje

```python
def curva_aprendizaje(model, train_sizes):
    train_errors = []
    test_errors = []

    for size in train_sizes:
        # incrementa el train-set
        X_train_sub = Xtrain.iloc[:size]
        y_train_sub = ytrain.iloc[:size]
        model.fit(X_train_sub, y_train_sub)
        y_train_pred = model.predict(X_train_sub)
        y_test_pred = model.predict(Xtest)

        mse_train = metrics.mean_squared_error(y_train_sub, y_train_pred)
        mse_test = metrics.mean_squared_error(ytest, y_test_pred)

        train_errors.append(mse_train)
        test_errors.append(mse_test)

    return train_errors, test_errors
```

```python
# (a, b, n) crea n numero esquiespaciados entre a y b
# numeros int
train_sizes = np.linspace(100, len(Xtrain)*0.9, 10, dtype=int)

DT = DecisionTreeRegressor(random_state=100)
knn = KNeighborsRegressor(n_neighbors=5)

DT_train_err, DT_test_err = curva_aprendizaje(DT, train_sizes)
knn_train_err, knn_test_err = curva_aprendizaje(knn, train_sizes)
```

```python
plt.figure(figsize=(8,4))

plt.plot(train_sizes, DT_train_err, label="DT Train MSE")
plt.plot(train_sizes, DT_test_err, label="DT Test MSE")

plt.xlabel("Tamaño del set de entrenamiento")
plt.ylabel("Error (MSE)")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,4))

plt.plot(train_sizes, knn_train_err, label="KNN Train MSE")
plt.plot(train_sizes, knn_test_err, label="KNN Test MSE")

plt.xlabel("Tamaño del set de entrenamiento")
plt.ylabel("Error (MSE)")
plt.legend()
plt.grid()
plt.show()
```

## Bias and Variance
El bias es la diferencia entre el valor predicho y el real. Esta dada por:
$$\text{bias}^{2}=E[\hat{y}]-y$$
La varianza mide cuanto cambian las predicciones del modelo cuando se entrena con distintos datos del subconjunto. Esta dada por:
$$\text{variance}=E[(\hat{y}-E[\hat{y}])^{2}]$$

El modelo se contruye a base distintso set de entrenamiento, producto de la recombinacion del set de entrenameinto original, pero usa siempre el mismo set de prueba original para evaluar. La esperanza es un promedio tomado entre muchas repeticiones.

Referencias: https://www.geeksforgeeks.org/machine-learning/bias-vs-variance-in-machine-learning/

```python
def bias_variance(model, Xtrain, ytrain, Xtest, ytest, n_runs=10):
    """
    input: modelo de ML, features y labels del set de ebtrenamiento y prueba, numero de iteraciones
    
    Cada numero de interacion se ocupa como una semilla para generar una division aleatoria de set de
    entrenamiento y prueba, considerando que el 20% del dataset es de prueba. Para cada iteracion se
    genera el modelo elegido con los datos de entranamiento dados y se obtiene la prediccion para el 
    set de prueba original (Xtest). Las predicciones se almacenan en una lista, la cual se conivierte
    en un array de dimensiones n x m, siendo n el numero de modelos (o iteraciones) y m el numero de 
    puntos en el Xtest. Luego se calcula el promedio de predicciones par cada punto y con ello se calcula
    el bias², variance y funcion de costo.
    
    output:bias², variance y funcion de costo.
    """
    preds = []

    for i in range(n_runs):
        Xt, _, yt, _ = train_test_split(Xtrain, ytrain, test_size=0.2, random_state=i)
        model.fit(Xt, yt)
        model_pred = model.predict(Xtest)
        preds.append(model_pred)

    # array con forma (numero de modelos, puntos en el Xtest)
    preds = np.array(preds)

    # Promedia predicciones verticalmente
    # obtiene el promedio de prediccion para cada punto del Xtest
    # es la esperanza de la prediccion
    mean_pred = preds.mean(axis=0)

    # Bias²: promedio del error entre la predicción promedio y el valor real
    #bias2 = np.mean((mean_pred - ytest.values) ** 2)
    bias2 = ((mean_pred -ytest)**2).mean()

    # Varianza: variabilidad entre predicciones individuales
    #variance = np.mean(np.var(preds, axis=0))
    variance = preds.var(axis=0).mean()

    f_costo = bias2 + variance

    return bias2, variance, f_costo
```

```python
DT = DecisionTreeRegressor(random_state=100)
knn = KNeighborsRegressor(n_neighbors=5)

bias2_DT, var_DT, f_costo_DT = bias_variance(DT, Xtrain, ytrain, Xtest, ytest)
bias2_knn, var_knn, f_costo_knn = bias_variance(knn, Xtrain, ytrain, Xtest, ytest)

print("DT:")
print("Bias²:", bias2_DT)
print("Var:", var_DT)
print("Funcion de costo:", f_costo_DT)
print(" ")
print("KNN:")
print("Bias²:", bias2_knn)
print("Var:", var_knn)
print("Funcion de costo:", f_costo_knn)
```
